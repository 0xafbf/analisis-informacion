```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python

def f(t):
    return 3 * np.cos(2.5*t) + 8 * np.sin(4*t) + np.cos(t+2) + 0.2 * np.cos(8*t)

samples = 50000
time = 30

t = np.linspace(0,time,samples)

ft = f(t)

fft = np.fft.rfft(ft)
t_fft = np.fft.rfftfreq(samples, time/samples)

plt.plot(t_fft, np.absolute(fft))
    
```
