```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg
```

```python
csv = np.genfromtxt('datos.csv', delimiter=',')
acc = csv[1:]

# me invente este sample rate
sample_rate = 60


datos_mag = linalg.norm(acc, axis=1)
# restar acc gravedad
# acc_norm = acc[:, 2]-9.8
# acc_norm = acc[:,0]
acc_norm = datos_mag


#tomaar muestras relevantes 
acc_norm = acc_norm[1560:2400]

t_acc = np.linspace(0, acc_norm.size/sample_rate, acc_norm.size)
#print(acc_norm)


# fourier
acc_fft = np.fft.rfft(acc_norm)

#freq_fft
#freq_fft = np.fft.rfftfreq(acc_norm.size, d=1./sample_rate)

freq_fft = np.linspace(0, sample_rate/2, acc_fft.size)
#freq_fft = freq_fft/sample_rate * 2

acc_fft = acc_fft * (np.heaviside(freq_fft - 5, 1))



plt.figure()
plt.plot(t_acc, acc_norm)

plt.figure()
plt.plot(freq_fft, np.absolute(acc_fft))

plt.figure()
plt.plot(t_acc, np.fft.irfft(acc_fft))

# plt.figure()
# plt.plot(datos[:,0])
# plt.plot(datos[:,1])
# plt.plot(datos[:,2])
# plt.figure()
# plt.plot(datos_mag)

# vel_x = np.cumsum(acc[:,0])
# vel_y = np.cumsum(acc[:,1])
# vel_z = np.cumsum(acc[:,2]-9.8)

```
